The main purpose of this notebook to show how metrics, optimzers and the application of regularization techniques impact a model's performance.  Overfitting occurs when you achieve a good fit of your model on the training data but the model doesn't generalize well on new, unseen data.  A model that is overfitting tries to learn too well the details and the noise of the training data which results in poor performance on the unseen data.  An example of overfitting is when the model performs exceptionally well on the training data but did not perform well on making predictions on unseen data.  We can identify overfitting by using performance metrics, such as loss or accuracy. Usually, the validation metric stops improving after a certain number of epochs and begins to decrease afterward. The training metric continues to improve because the model seeks to find the best fit for the training data.  A way to address overfitting in deep neural networks is to apply regularization techniques.  Regularization techniques make slight modifications to the learning algorithm such that the model generalizes better and improves the modelâ€™s performance on the unseen data as well.

I will train the model with the regularizers L1, L2, Dropout, and Early stopping.  In addition, I will train the models with the optimizers:  Adam, SGD, and RMSProp. 

I will create a table noting the model performance for each metric.  In the table I will note the optimal hyperparameters, such as number of epochs, best regularizer, the best optimizer.

```python
def plot_metric(history, metric1, metric2):
  train_metrics = history.history[metric1]
  val_metrics = history.history['val_'+metric1]
  epochs = range(1, len(train_metrics) + 1)
  # plot accuracy
  fig, (ax1, ax2) = plt.subplots(2,figsize=(9,9))
  ax1.plot(epochs, train_metrics)
  ax1.plot(epochs, val_metrics)
  ax1.set_title('Training and Validation '+ metric1)
  ax1.set_xlabel("Epochs")
  ax1.set_ylabel(metric1)
  ax1.legend(["train_"+metric1, 'val_'+metric1])
  # plot loss
  train_metrics2 = history.history[metric2]
  val_metrics2 = history.history['val_'+metric2]
  epochs2 = range(1, len(train_metrics2) + 1)
  ax2.plot(epochs2, train_metrics2)
  ax2.plot(epochs2, val_metrics2)
  ax2.set_title('Training and Validation '+ metric2)
  ax2.set_xlabel("Epochs")
  ax2.set_ylabel(metric2)
  ax2.legend(["train_"+metric2, 'val_'+metric2])
  fig.tight_layout(pad=2.0)
  plt.show()
```

### Regularizer: Dropout
The Dropout technique is used during the training of the model.  At every iteration, the technique randomly selects nodes based on the percentage identified and removes them along with all of their incoming and outgoing connections.  Each iteration has a different set of nodes and this results in a different set of outputs. The probability of choosing how many nodes should be dropped is the hyperparameter of the dropout function.  The dropout can be applied to both the hidden layers as well as the input layers.  Dropout is usually preferred when we have a large neural network structure in order to introduce more randomness.
A good value for dropout in a hidden layer is between 0.5 and 0.8. Input layers use a larger dropout rate, such as of 0.8.

```python
#from keras.layers.core import Dropout
from tensorflow.keras.layers import Dropout
#from tensorflow.keras.regularizers import l2
from keras.constraints import max_norm
```

```python
# def define_model():
#   model = Sequential()
#   Dense(64, activation="relu", input_shape=(8,)),
#   Dropout(0.8),
#   Dense(128, activation="relu"),
#   Dropout(0.5),
#   Dense(128, activation="relu"),
#   Dropout(0.5),
#   Dense(128,  activation="relu"),
#   Dropout(0.5),
#   Dense(64, activation="relu"),
#   Dropout(0.5),
#   Dense(64, activation="relu"),
#   Dropout(0.5),
#   Dense(64, activation="relu"),
#   Dropout(0.5),
#   Dense(1, activation='sigmoid')
#   #compile model
#   model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#   #print(model.summary())
#   return model
def create_regularized_model(rate1, rate2):
    model_d = Sequential([
        Dense(64, activation="relu", input_shape=(8,)),
        Dropout(rate1),
        Dense(128, activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(128, activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(128,  activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(64, activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(64, activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(64, activation="relu",kernel_constraint=max_norm(3)),
        Dropout(rate2),
        Dense(1, activation='sigmoid')
    ])
    return model_d
```

```python
# Re-build the model with weight decay and dropout layers
model_d = create_regularized_model(0.8, 0.5)
```

```python
model_d.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['accuracy']
)
```

### Early Stopping
This technique stops the training when generalization error increases.
Here, monitor denotes the quantity that needs to be monitored and â€˜val_errâ€™ denotes the validation error.

Patience denotes the number of epochs with no further improvement after which the training will be stopped. For better understanding, letâ€™s take a look at the above image again. After the dotted line, each epoch will result in a higher value of validation error. Therefore, 5 epochs after the dotted line (since our patience is equal to 5), our model will stop because no further improvement is seen.



```python
from keras.callbacks import EarlyStopping

#EarlyStopping(monitor='val_err', patience=5)

#callback = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=4, restore_best_weights=True)
custom_early_stopping = EarlyStopping(
    monitor='val_accuracy', 
    patience=8, 
    min_delta=0.001, 
    mode='max'
)
```

```python
# def create_model(): 
#   model_es = Sequential([
#     Dense(128, input_shape=(8,),  kernel_regularizer=regularizers.l2(0.01),activation='relu'), #input_dim=8, 
#     Dense(128, activation='relu'),
#     Dense(128, activation='relu'),
#     Dense(32, activation='relu'),
#     Dense(1, activation='sigmoid')
#   ])
  
#  return model_es

def create_model(): 
  model_es = Sequential([
      Dense(64, activation='relu', input_shape=(8,)),
      Dense(128, activation='relu'),
      Dense(128, activation='relu'),
      Dense(128, activation='relu'),
      Dense(64, activation='relu'),
      Dense(64, activation='relu'),
      Dense(64, activation='relu'),
      Dense(1, activation='sigmoid')])
    
  return model_es
```

```python
#create tensor.  Floattensor converts to float 
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test =  torch.LongTensor(y_test)
```

```python
### creating model w/Pytorch

class ANN_Model(nn.Module):
    def __init__(self, input_features=8, hidden1=20, hidden2=20, out_features=2):
        super().__init__()
        self.f_connected1 = nn.Linear(input_features, hidden1) #create a Dense layer
        self.f_connected2 = nn.Linear(hidden1, hidden2)
        self.out = nn.Linear(hidden2,out_features )
    def forward(self,x):  #forward propogation
        x = F.relu(self.f_connected1(x))
        x = F.relu(self.f_connected2(x)) 
        x= self.out(x) 
        return x  
```

```python
### instantiate ANN_model
torch.manual_seed(20)
model = ANN_Model()
```

```python
model.parameters
```

```python
# backward propogation-- define the loss function, define the optimizer
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
epochs=500
final_losses = []
for i in range(epochs): #it's going to run for this number of epochs
    i = i+1
    y_pred = model.forward(X_train)
    loss = loss_function(y_pred, y_train)
    final_losses.append(loss)
    if i % 10 ==1: #for every 10 epochs print something
         print("Epoch number: {} and the loss: {}".format(i,loss.item()))
    #reduce the loss
    optimizer.zero_grad()  
    loss.backward() #back propogation
    optimizer.step()
```
